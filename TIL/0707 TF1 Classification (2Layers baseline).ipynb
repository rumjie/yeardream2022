### Normalization

```python
#   [28 x 28] [1 x 784] . 
#    pixel  [0~255] [0~1] .

train_data = train_data.reshape(60000, 784) / 255.0 #(60000, -1)  
test_data = test_data.reshape(10000, 784) / 255.0
```

### 2. Build the model

```python
# import tensorflow as tf

import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2' # https://stackoverflow.com/questions/35911252/disable-tensorflow-debugging-information
tf.logging.set_verbosity(tf.logging.ERROR)
```

```python
#   (placeholder)  

X = tf.placeholder(tf.float32, [None, 28*28]) #   
# of batch data, # of features(columns) ==  784 ]
#None        
Y = tf.placeholder(tf.float32, [None, 10]) # 0~9 ==  10   
```

```python
#  Parameter Theta Variable 
#   ,  
#        

W1 = tf.Variable(tf.random_normal([784, 256], stddev=0.01)) #256 =   , weight  
#random_normal =   ,    
L1 = tf.nn.relu(tf.matmul(X, W1)) # hidden layer,  = None * 256
```

```python
W2 = tf.Variable(tf.random_normal([256, 256], stddev=0.01)) #   &    
L2 = tf.nn.relu(tf.matmul(L1, W2))
```

```python
W3 = tf.Variable(tf.random_normal([256, 10], stddev=0.01)) #   &   
#This is output layer
model = tf.matmul(L2, W3) #    
#softmax : tf.nn.softmax()   
```

### 3. Set the criterion

```python
#cost = tf.losses.mean_squared_error(Y, model) # for Regression

cost = tf.losses.softmax_cross_entropy(Y, model) # for Classification, "cross-entropy" after "softmax"
#  
optimizer = tf.train.AdamOptimizer(0.001).minimize(cost) # Select optimizer & connect with cost function (recommended start : "Adam")
```

### 4. Train the model

```python
init = tf.global_variables_initializer() # Initialize all global variables (Parameter Theta)
sess = tf.Session() #     ( )
sess.run(init) 
```

```python
print(sess.run(W1))
sess.run(W1).shape #run   theta   
```

```python
train_data.shape
```

```python
# Gradient descent        (Batch size ==      )

batch_size = 100 #-> 1 epoch 600  

total_batch = int(len(train_data) / batch_size)

print(total_batch)
```

```python
#import tqdm 
#for epoch in tqdm.notebook.tqdm(range(15)):
#for    progress   
#15  = 600*15  
```

```python
import tqdm 
```

```python

for epoch in range(15):
    
    
    total_cost = 0 #  epoch       .
    batch_idx = 0 #  batch     index    .
    
    for i in range(total_batch): # iterate over # of batches
        
        # Training data(60000) batch_size(100)     .
        batch_x = train_data[ batch_idx : batch_idx + batch_size ] # 0:100 
        batch_y = train_label[ batch_idx : batch_idx + batch_size ] #100 
        
        sess.run(optimizer, feed_dict={X: batch_x, Y: batch_y}) # (feed) (dict)
        #X, Y placeholder    
        
        #  batch    Cross-entropy  total_cost . (epoch    .)
        batch_cost = sess.run(cost, feed_dict={X: batch_x, Y: batch_y}) # ! cost   
        #Error of training, cost of batch 
        total_cost = total_cost + batch_cost #sum of 600 errors
        
        #  for loop    index  batch_size(100)  .
        batch_idx += batch_size
    
    
    # ( epoch (= 600)  ) training data  Cross-entropy  .
    training_cost = total_cost / total_batch #error of this epoch
    
    # ( epoch  ) test data  Cross-entropy  .
    test_cost = sess.run(cost, feed_dict={X: test_data, Y: test_label})  
    
    
    print('Epoch: {}'.format(epoch + 1), 
          '|| Avg. Training cost = {:.3f}'.format(training_cost), 
          '|| Current Test cost = {:.3f}'.format(test_cost))

print('Learning process is completed!')

#  !   
```

###   (tf.argmax / tf.equal / tf.cast / tf.reduce_mean )

```python
# tf.argmax([0.1 0 0 0.7 0 0.2 0 0 0 0]) -> 3 (   index return)

is_correct = tf.equal(tf.argmax(model, 1), tf.argmax(Y, 1)) # model : , Y :  
#equal =  true,  false Return 
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32)) #  (type-"cast") ,  (reduce) (mean) 
```

```python
# 10,000 Test data     .

print(' :', sess.run(accuracy,
                        feed_dict={X: test_data,
                                   Y: test_label}))
```
